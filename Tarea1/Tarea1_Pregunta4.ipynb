##Función

```python
import math

def f(x):
    sum = 3.0 * x
    coseno = 4.0 * math.cos(2.0 * x)
    res = math.sin(sum + coseno)
    return res
```

##Gráfico

```python
# Graficar función con todos las raices posible rapidamente
import matplotlib.pyplot as plt
import math
import numpy as np

ejeX=[]; ejeY=[]; solo0 = []

x = 0.0
numPi = float(math.pi)

while x <= numPi * 2:
    y = f(float(x))
    ejeX.append(x)
    ejeY.append(y)
    solo0.append(0)
    x += numPi / 10

plt.plot(ejeX,ejeY)
plt.plot(ejeX,solo0)
plt.legend(["Función", "y = 0"])

plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

```python
# Graficar función lo mas precisa posible
import matplotlib.pyplot as plt
import math

ejeX=[]; ejeY=[]; solo0 = []

x = 0.0
numPi = float(math.pi)

while x <= numPi * 2:
    y = f(float(x))
    ejeX.append(x)
    ejeY.append(y)
    solo0.append(0)
    x += 0.000001


plt.plot(ejeX,ejeY)
plt.plot(ejeX,solo0)
plt.legend(["Función", "y = 0"])

plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

###Comentario.

Se puede observar por medio del gráfico que la función $y = sin(3x+ 4 cos(2x))$ tiene una increíble cantidad de $10$ raíces en el intervalo $x = [0, 2π]$. Esto puede deberse a que es una función de seno, es decir, una función trigonométrica que tienden a mantenerse cerca de $y=0$ pasando por $y,-y$ de un lado a otro.

Además, es una función trigonométrica **compuesta**, $f(x) = sin(x)$, $g(x) = 3x + 4cos(x)$. $y = f(g(x))$ podemos decir que ambas funciones que la componen son trigonométricas, $seno$ y $coseno$ son conocidos por tener innumerables raíces, por eso no es anormal que tenga tantas raíces en un pequeño intervalo.

Debido a que es una función seno y coseno, sus raices se encuentran en terminos de $\pi$, es decir, son números indefinidos con decimales infinitos.

##Método de Busqueda Incremental

```python
import math

def BusquedaIncremental(x):
    X=[]; raices =[]
    numPi = float(math.pi)

    while x <= numPi * 2:
        y = f(float(x))
        if y == 0.0:
            raices.append(x)
        x += numPi / 100
        X.append(x)
    return X, raices

all, raiz = BusquedaIncremental(0.0)

print(f"Se evaluaron {len(all)} números como posibles raices.")
print(f"raices encontradas = {raiz}")
```

##Método Bisección

```python
import math

def PuntoMedio(a,b):
    m = a + b
    m /= 2.0
    return m

# Busca Raíz exacta
def BiseccionMethodPrecise(a,b):
    raices = []
    X = []
    Y = []
    for i in range(80):
        m = PuntoMedio(a,b)
        fM = f(float(m))
        X.append(m)
        Y.append(fM)
        if fM == 0:
            raices.append(m)
        if fM * f(float(a)) < 0:
            b = m
        else:
            a = m
    return X, Y,raices

posiblesBiseccion, resultPosiblesRaicesBiseccion,resultadosBiseccion = BiseccionMethodPrecise(3.0,4.0)
print(f"números evaluados = {posiblesBiseccion}")
print(f"Se evaluaron {len(posiblesBiseccion)} números como posibles raices.")
print(f"raices encontradas = {resultadosBiseccion}")
```

###Comentario.

Cuando se busca con el método de Bisección un $x$ que de precisamente una raíz $f(x) = y = 0$, no encuentra ninguna raíz dentro de las primeras $80$ iteraciones. 

Esto puede deberse a que las raíces exactas, no existen, las raices de esta ecuación tienen infinitos decimales, por eso con el metodo de Bisección no encontraremos nunca una raíz exacta, pero como usamos python, en algun momento aproximara el resultado de $f(x)$ a 0, ya que en algun punto ha de aproximarlo, cuando la diferencia es lo suficientemente pequeña. Pero seran muchisimas iteraciones las necesarias para llegar a ello.

##Método Regula Falsi

```python
import math

# Busca Raíz exacta
def RegulaFalsiMethodPrecise(a,b):
    raices = []
    X = []
    Y = []
    for i in range(80):
        fB = f(float(b))
        fA = f(float(a))
        xr = b
        num = fB * a 
        num -= fB * b
        denominador = fA - fB
        div = num / denominador
        xr -= div
        fXr = f(float(xr))
        if fXr==0:
            raices.append(xr)
        X.append(xr)
        Y.append(fXr)
        if fXr * f(float(b)) < 0:
            a = xr
        else:
            b = xr
        
    return X, Y,raices

posiblesRegulaFalsi, resultPosiblesRaicesRegulaFalsi,resultadosRegulaFalsi = RegulaFalsiMethodPrecise(3.0,4.0)
print(f"números evaluados = {posiblesRegulaFalsi}")
print(f"Se evaluaron {len(posiblesRegulaFalsi)} números como posibles raices.")
print(f"raices encontradas = {resultadosRegulaFalsi}")
```

```python
def esZero(num):
    if num > -0.00000001 and num < 0.00000001:
        return True
    return False
```

###Biseccion

```python
import math

# Busca aproximados a la raíz junto a la raíz
def BiseccionMethod(a,b):
    raices = []
    X = []
    Y = []
    for i in range(80):
        m = PuntoMedio(a,b)
        fM = f(float(m))
        X.append(m)
        Y.append(fM)
        if esZero(fM):
            raices.append(m)
        if fM * f(float(b)) < 0:
            a = m
        else:
            b = m
    return X, raices

posiblesBiseccion, resultadosBiseccion = BiseccionMethod(3.0,4.0)
print(f"números evaluados = {posiblesBiseccion}")
print(f"Se evaluaron {len(posiblesBiseccion)} números como posibles raices.")
print(f"raices aproximadas encontradas = {resultadosBiseccion}")
```

####Comentario

```python
import math

# Busca Raíz aproximada y la raíz
def RegulaFalsiMethod(a,b):
    raices = []
    X = []
    Y = []
    for i in range(80):
        fB = f(float(b))
        fA = f(float(a))
        xr = b
        num = fB * a 
        num -= fB * b
        denominador = fA - fB
        div = num / denominador
        xr -= div
        fXr = f(float(xr))
        if esZero(fXr):
            raices.append(xr)
        X.append(xr)
        Y.append(fXr)
        if fXr * f(float(b)) < 0:
            a = xr
        else:
            b = xr
        
    return X, raices

posiblesRegulaFalsi, resultadosRegulaFalsi = RegulaFalsiMethod(0.0,2.0 * math.pi)
print(f"números evaluados = {posiblesRegulaFalsi}")
print(f"Se evaluaron {len(posiblesRegulaFalsi)} números como posibles raices.")
print(f"raices aproximadas encontradas = {resultadosRegulaFalsi}")
```
