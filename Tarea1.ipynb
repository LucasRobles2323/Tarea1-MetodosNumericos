##Código.

```python
import numpy as np
import math
import pandas as pd

def DivideYpromedia(initial, cantMax, a):
    I = [] # Arreglo de initial
    Ea = [] # Arreglo de Errores aproximados
    Eabs = [] # Arreglo de Errores absolutos
    Vapp = [] # Valores aproximados de la raíz obtenidos
    
    trueValor = math.sqrt(a) # Valor real

    for i in range(0, cantMax):
      aux = initial

      # Calculo de Divide y Promedia-----
      initial = float(aux + (a / aux)) # Calcula el numerador
      initial = float(initial / 2) # Lo divide en 2, se completa la división
      # ---------------------------------

      # Calculo----------------------------------------------------
       # El valor app queda como initial
      error_relav = (trueValor - initial)/trueValor

      # Llenado de arreglos
      Ea.append(error_relav)
      Eabs.append(abs(error_relav))
      Vapp.append(initial)

      #------------------------------------------------------------

      df = pd.DataFrame({'Error relativo aproximado   |': Ea,
                         'Error relativo absoluto   |': Eabs,
                         'Valor aproximado': Vapp })
      
    I.append(initial) # Arreglo de initials, para corroborar
    return df


# Zona de Input---------------------------------------------------------------
a = float(input("Ingrese un número positivo para su raíz: "))
initial = a / 2.0

cantMax = int(input("Ingrese cantidad máxima de iteraciones: "))

df = DivideYpromedia(initial, cantMax, a)

print(f"La raiz vale: {math.sqrt(a)}")
print ("\n----------------------------------------------------------")
df
```

##Fórmula:

>   $\sqrt{Q} = \sqrt{A^2 - b} = A + \frac{{b}}{2A} - \frac{(\frac{{b}}{2A})^2}{2(A + \frac{{b}}{2A})} $

**Al descomponer esta expresion:**
> $A + \frac{{b}}{2A} - \frac{(\frac{{b}}{2A})^2}{2(A + \frac{{b}}{2A})}$  (1)

**Se obtiene los componentes:**
- $C1$ = $A$
- $C2$ = $\frac{{b}}{2A}$
- $C31$ = $(\frac{{b}}{2A})^2$ = $($C2$)^2$
- $C32$ = ${2(A + \frac{{b}}{2A})}$ = $2($C1$+$C2$)$

**Se plantea la misma expresion (1) en terminos de componentes:**
> $C1 + C2 -\frac{{C31}}{C32}$

##Código

```python
import numpy as np
import math
import pandas as pd

def DivideYpromedia(initial, cantMax, a):
    I = [] # Arreglo de initial
    Ea = [] # Arreglo de Errores aproximados
    Eabs = [] # Arreglo de Errores absolutos
    
    trueValor = math.sqrt(a) # Valor real

    for i in range(0, cantMax):
      aux = initial 

      # Calculo de Divide y Promedia-----
      initial = float(aux + (initial / aux)) # Calcula el numerador
      initial = float(initial / 2) # Lo divide en 2, se completa la división
      # ---------------------------------

      print(str(initial))

      # Calculo----------------------------------------------------
       # El valor app queda como initial
      error_relav = (trueValor - initial)/trueValor

      # Llenado de arreglos
      Ea.append(error_relav)
      Eabs.append(abs(error_relav))

      #------------------------------------------------------------

      df = pd.DataFrame({'Error relativo aproximado   |': Ea,
                         'Error relativo absoluto': Eabs })
      
    I.append(initial) # Arreglo de initials, para corroborar
    return df
```

```python
import math

# Funcion para buscar el cuadrado perfecto mas cercano al numero ingresado
def cuadradoPerfecto(num):
  while(num > 0):
    # Calcula la raiz y obtiene la parte decimal
    sqrt = math.sqrt(num)
    decimal = sqrt - math.trunc(sqrt)
    
    # 0.0 es raiz perfecta, caso contrario disminuyo el numero en una unidad
    if(decimal == 0.0):
      break
    num -= 1
  
  return num
```

```python
import math

# Funcion para calcular la raiz cuadrado de un numero por metodo MB
def metodoMB(Q):
  
  # Verifica si ingreso un numero negativo
  if(Q < 0):
    print(str(Q) + " es un numero negativo. Se cancela la operacion\n")
    return
  
  # Corresponde A^2
  A = cuadradoPerfecto(Q)
  
  # Pregunta si el numero ingresado es un cuadrado perfecto
  if(Q == A):
    print(str(A) + " es un numero cuadrado perfecto\n")
    return
  
  # Obtengo la variable 'b' y calculo la raiz cuadrada del numero ingresado por componentes
  #
  # Formula: A + b/(2A) - [(b/(2A))^2]/[2(A + b/(2A))]
  #
  b = Q - A
  C1 = math.sqrt(A) # A
  C2 = b/(2*(C1)) # b/(2A)
  C31 = pow(C2, 2) # [(b/(2A))^2]
  C32 = 2*(C1 + C2) # [2(A + b/(2A))]

  # Junto cada componentes con respecto a la formula y obtengo el valor app. de la raiz
  valor_formula = C1 + C2 -(C31/C32)
  
  # Valor real de la raiz
  real = math.sqrt(Q)
  print("\nValor real de " + str(Q) + " = " + str(real))
  print("Valor aproximado de " + str(Q) + " = " + str(valor_formula))
  
  # Calculo el error relativo
  error = (abs(real - valor_formula)/real * 100)
  print("Error relativo = " + str(error) + "%\n")

  # DIVIDE Y PROMEDIA...
  initial = Q / 2

  #Cantidad requerida de iteraciones
  cantMax = 5

  df = DivideYpromedia(initial, cantMax, Q)

  print ("--------------------------------------------")
  print(df)

# INICIA AQUI...
for x in range (1,4):
  numero = int(input("\nIteracion n°" + str(x) + "\nNumero = "))
  metodoMB(numero)
```

```python
import numpy as np
import pandas as pd

def func(x):
  g = 9.81
  m = 80
  v = 36
  t = 4
  app = (np.sqrt((g*m)/x)*np.tanh(np.sqrt((g*x)/m)*t))-v
  return app

def valorC(a,b):
  resultado = (a+b)/2
  return resultado

def error(valor_aprox, valor_vdd):
  return abs((valor_vdd - valor_aprox)/valor_vdd)
```

###Método de Busqueda Incremental

```python
import math

def BusquedaIncremental(x, delta):
    Intervalos = [] # Guardara todos los intervalos encontrados por el Método
    
    # Inicializando variables que marcan el intervalo
    a = 0
    b = delta

    while b <= float(math.pi) * 2:
        fA = f(float(a))
        fB = f(float(b))
        if fB * fA < 0:
            Intervalos.append((a,b))
        a += delta
        b += delta
    return Intervalos

intervalos = BusquedaIncremental(0.0, float(math.pi) / 10.0)

for i in range(len(intervalos)):
    print(intervalos[i])
```

###Método Bisección

```python
import math

def PuntoMedio(a,b):
    m = a + b
    m /= 2.0
    return m

# Busca Raíz exacta
def BiseccionMethodPrecise(a,b):
    raices = []
    X = []
    Y = []
    for i in range(80):
        m = PuntoMedio(a,b)
        fM = f(float(m))
        X.append(m)
        Y.append(fM)
        if fM == 0.0:
            raices.append(m)
        if fM * f(float(a)) < 0:
            b = m
        else:
            a = m
    return X, Y,raices

def mainBiseccion(a,b, verbose=True):
    posiblesBiseccion, resultPosiblesRaicesBiseccion,resultadosBiseccion = BiseccionMethodPrecise(a,b)
    if verbose:
        print(f"Intervalo evaluado por el método Bisección: [{a}, {b}]")
        print(f"Números evaluados = {posiblesBiseccion}")
        print(f"Se evaluaron {len(posiblesBiseccion)} números como posibles raices.")
        print(f"Raices encontradas = {resultadosBiseccion}")
    return posiblesBiseccion, resultPosiblesRaicesBiseccion
```

###Método Regula Falsi

```python
import math

# Busca Raíz exacta
def RegulaFalsiMethodPrecise(a,b):
    raices = []
    X = []
    Y = []
    for i in range(80):
        fB = f(float(b))
        fA = f(float(a))
        xr = b
        num = fB * a 
        num -= fB * b
        denominador = fA - fB
        div = num / denominador
        xr -= div
        fXr = f(float(xr))
        if fXr==0:
            raices.append(xr)
        X.append(xr)
        Y.append(fXr)
        if fXr * f(float(b)) < 0:
            a = xr
        else:
            b = xr
        
    return X, Y,raices

def mainRegulaFalsi(a,b,verbose=True):
    posiblesRegulaFalsi, resultPosiblesRaicesRegulaFalsi = RegulaFalsiMethodPrecise(a,b)
    if verbose:
        print(f"Intervalo evaluado por el método Regula-Falsi: [{a}, {b}]")
        print(f"Números evaluados = {posiblesRegulaFalsi}")
        print(f"Se evaluaron {len(posiblesRegulaFalsi)} números como posibles raices.")
        print(f"Raices encontradas = {resultadosRegulaFalsi}")
    return posiblesRegulaFalsi, resultPosiblesRaicesRegulaFalsi
```

```python
def esZero(num):
    if num > -0.00000001 and num < 0.00000001:
        return True
    return False
```

####Biseccion

```python
import math

# Busca aproximados a la raíz junto a la raíz
def BiseccionMethod(a,b):
    raices = []
    X = []
    Y = []
    for i in range(80):
        m = PuntoMedio(a,b)
        fM = f(float(m))
        X.append(m)
        Y.append(fM)
        if esZero(fM):
            raices.append(m)
        if fM * f(float(b)) < 0:
            a = m
        else:
            b = m
    return X, raices

posiblesBiseccion, resultadosBiseccion = BiseccionMethod(3.0,4.0)
print(f"números evaluados = {posiblesBiseccion}")
print(f"Se evaluaron {len(posiblesBiseccion)} números como posibles raices.")
print(f"raices aproximadas encontradas = {resultadosBiseccion}")
```
